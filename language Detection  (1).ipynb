```python
data = pd.read_csv('/content/drive/MyDrive/language.csv')
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tokenizer setup
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(data['Text'])

# Convert to sequences
sequences = tokenizer.texts_to_sequences(data['Text'])

# Padding
max_len = 100
padded = pad_sequences(sequences, maxlen=max_len, padding='post')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(data['language'])
```

```python
# checking label assign to the languages
for i, class_name in enumerate(label_encoder.classes_):
    print(f"{i}: {class_name}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(padded, labels, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=64, input_length=max_len))
model.add(LSTM(64))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=15, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("LSTM Model Accuracy:", accuracy)
```

```python
import pickle

# Save the trained Naive Bayes model
with open("nb_model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save the CountVectorizer
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(cv, f)

# Save the LabelEncoder
with open("label_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder, f)
```

```python
import gradio as gr
import pickle
import numpy as np

# Load all pickle files
with open("nb_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("vectorizer.pkl", "rb") as f:
    cv = pickle.load(f)

with open("label_encoder.pkl", "rb") as f:
    label_encoder = pickle.load(f)

# Define prediction function
def predict_language(text):
    try:
        vec = cv.transform([text]).toarray()
        pred = model.predict(vec)
        language = label_encoder.inverse_transform(pred)
        return language[0]
    except Exception as e:
        return f"‚ùå Error: {e}"

# Gradio interface
iface = gr.Interface(
    fn=predict_language,
    inputs=gr.Textbox(lines=2, placeholder="Enter a sentence..."),
    outputs="text",
    title="üåê Language Detection with Naive Bayes",
    description="Enter a sentence and this model will detect the language using Naive Bayes algorithm."
)

# Launch the app
iface.launch(share=True)

```
